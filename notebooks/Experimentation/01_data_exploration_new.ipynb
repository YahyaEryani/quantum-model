#Exploratory Data Analysis
In this section, we will perform exploratory data analysis by visualizing and analyzing the data to identify any patterns, trends, or relationships between different features. We will also identify any potential issues or anomalies in the data.

```python
#in this figure we see that the heat map shows linear correlation for example, the same logic applied for the resr of the values
# Calculate Spearman correlation matrix
spearman_corr = data_clean.corr(method='spearman')

# Create heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(spearman_corr, cmap="coolwarm", annot=True, fmt='.2f')
plt.title('Spearman Correlation Matrix')
plt.show()
```

#Splitting The Dataset

In this section, we will split the dataset into three parts: training, validation, and test sets. The training set will be used to train the XGBoost model, the validation set will be used to tune the model's hyperparameters and prevent overfitting, and the test set will be used to evaluate the model's performance on unseen data. We will be using a 60% training, 20% validation, and 20% test split. Once the data is split, we will save each set to a CSV file for further use.

```python
# Split the data into 70% training, 15% validation, and 15% testing
train_data, test_data = train_test_split(data_clean, test_size=0.15, random_state=42, shuffle=True, stratify=data_clean['class_label'])
train_data,val_data = train_test_split(train_data, test_size=0.177, random_state=42, shuffle=True, stratify=train_data['class_label'])

# Save the data to pickle files
train_data.to_pickle('training_data_drpped_column.pkl')
val_data.to_pickle('validation_data_drpped_column.pkl')
test_data.to_pickle('testing_data_drpped_column.pkl')

#make sure the probability distribution is the same, add stratify to make the distribution homogenous
#suffle=true, stratify the test 
#adysh rah ya3mel adaptation aa kol observation because we have enough observation
```
